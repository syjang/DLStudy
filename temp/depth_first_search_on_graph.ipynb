```python
import pickle
import networkx as nx
```

```python
# to read it back
with open ('H{}'.format("24802"), 'rb') as fp:
    G = pickle.load(fp)
```

```python
G
```

```python
list(nx.dfs_edges(G, source = 'black', depth_limit= 1))
```

```python
#page rank on the Graph: the propbability represents the priority(or importance) of each word

pr = nx.pagerank(G, alpha=0.85)
pr_sorted_keys = sorted(pr, key=pr.get, reverse=True)
for i in pr_sorted_keys:
    print(i, pr[i])
```
